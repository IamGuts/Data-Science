### Метод подбора
y = 0.4*x + 3 + noise

```python
from ipywidgets import interactive
a = np.linspace(-5,5,50)
noise = np.random.uniform(-5,5, size=(50,))
d = .4 * a + 3 + noise

def f(m, b):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.scatter(a,d)
    plt.plot(x, m * x + b, color='red')
    plt.ylim(-3, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '400px'
interactive_plot
```

## Расчет beta коэффициентов с помощью матричной математики. Расчет потерь - метод наименьших квадратов

#### Dataset

```python
X_1 = np.linspace(1,30, 150)
X_2 = X_1 + 10
Y = X_1 * 2  + X_2 + np.random.rand()
```

```python
df = pd.DataFrame({'X_1': X_1, 'X_2': X_2, 'Y': Y})
df.head()
```

#### Beta coeffs: <i><font color='red'> np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y)) </font></i>

```python
X = df.drop('Y', axis=1)
Y = df['Y']
```

```python
weights = np.dot((np.linalg.inv(np.dot(X.T,X))), np.dot(X.T,Y))
```
